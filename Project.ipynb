```python
M = Matrix(
    [
        [m1, 0, 0, 0, 0, 0],
        [0, m2, 0, 0, 0, 0],
        [0, 0, m3, 0, 0, 0],
        [0, 0, 0, m4, 0, 0],
        [0, 0, 0, 0, m5, 0],
        [0, 0, 0, 0, 0, m6],
    ]
)
K = Matrix(
    [
        [k1 + k2, -k2, 0, 0, 0, 0],
        [-k2, k2 + k3, -k3, 0, 0, 0],
        [0, -k3, k3 + k4, -k4, 0, 0],
        [0, 0, -k4, k4 + k5 + k6, -k5, -k6],
        [0, 0, 0, -k5, k5, 0],
        [0, 0, 0, -k6, 0, k6],
    ]
)
C = Matrix(
    [
        [c1 + c2, -c2, 0, 0, 0, 0],
        [-c2, c2 + c3, -c3, 0, 0, 0],
        [0, -c3, c3 + c4, -c4, 0, 0],
        [0, 0, -c4, c4 + c5 + c6, -c5, -c6],
        [0, 0, 0, -c5, c5, 0],
        [0, 0, 0, -c6, 0, c6],
    ]
)
```

```python
omega = symbols("omega", positive=True)
detA = (
    14159710231981598304 * omega**12
    - (621719671149961893345367825396371 * omega**10) / 10995116277760
    + (35050848776533742656295507205498664899695502531 * omega**8)
    / 483570327845851669882470400
    - (75311578242617871070629674718747481141348436261918044721138687 * omega**6)
    / 2177807148294006166165597487563316553318400
    + (
        28996427679362925195085010662455197722052597135734893085824058172558938383
        * omega**4
    )
    / 4789048565205902682369834459844716198808559756823756800
    - (
        431158708267470257041787205902251949118437704530015312064916987461562820020692696937
        * omega**2
    )
    / 1053122916685571866979180276836704323188950954005491112543109775360
    + 14270296780537614628241279690607715985190755552271690254862890962536224913529455991
    / 1684996666696914987166688442938726917102321526408785780068975640576
)
slv = solve(detA)
slv
```

```python
u1, u2, u3, u4, u5, u6 = symbols("u_1 u_2 u_3 u_4 u_5 u_6")
U = Matrix([[1], [u2], [u3], [u4], [u5], [u6]])
for i, omega in enumerate(slv):
    eqns = list((K - omega**2 * M) * U)
    eqns.pop(1)
    slv_u = solve(eqns)
    print(
        f"mode shape {i + 1}:",
        [
            1,
            slv_u[u2],
            slv_u[u3],
            slv_u[u4],
            slv_u[u5],
            slv_u[u6],
        ],
    )
```
